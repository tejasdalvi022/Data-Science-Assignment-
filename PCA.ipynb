### Normalizing the values

```python
from sklearn.preprocessing import scale 
```

```python
df_scale = df.values
df_scale
```

```python
df_normal=scale(df_scale)
```

```python
df_normal
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca_values = pca.fit_transform(df_normal)
```

```python
pca_values
```

```python
pca = PCA(n_components=6)
pca_values = pca.fit_transform(df_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

## Transform the normal dataset into principal component

```python
# Set the number of components to the optimal number you determined
optimal_components = 2 
pca = PCA(n_components=optimal_components)

# Fit and transform the data
principal_components = pca.fit_transform(df_normal)

# Convert to DataFrame for easier handling
import pandas as pd
principal_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(optimal_components)])
```

```python
principal_df
```

## Kmeans clustering for normal data

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
fig=plt.figure(figsize=(8,4))
WCSS=[]
for i in range(1,11):
  model=KMeans(n_clusters=i).fit(df_normal)
  WCSS.append(model.inertia_)
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show
```

```python
clf=KMeans(n_clusters=3)
clf.fit(df_normal)
```

```python
y_names=clf.predict(df_normal)
```

```python
y_names
```

```python
clf.labels_
```

```python
md=pd.Series(y_names)
df['clust']=md
df.head()
```

```python
import sklearn
sklearn.metrics.silhouette_score(df_normal,clf.labels_)
```

```python
plt.style.use('classic')
print(plt.style.available)
```

```python
df.plot("Alcohol","Malic",c=clf.labels_,kind="scatter",s=50,cmap=plt.cm.coolwarm_r)
```

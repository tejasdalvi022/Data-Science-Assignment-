```python
# Calculate IQR
Q1 = df['LB'].quantile(0.25)
Q3 = df['LB'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[(df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

```python
# Replace outliers with median
median = df['LB'].median()
df['LB'] = np.where((df['LB'] < (Q1 - 1.5 * IQR)) | (df['LB'] > (Q3 + 1.5 * IQR)), median, df['LB'])
print(df)
```

## Statistical Summary:

```python
# Summary statistics
summary =df.describe()
print(summary)
```

```python
# Calculate IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Add IQR to the summary
summary.loc['IQR'] = IQR
print(summary)
```

```python
# Calculate mean, median, std, and IQR
mean = df.mean()
median = df.median()
std = df.std()
iqr = IQR

# Create a summary DataFrame
detailed_summary = pd.DataFrame({
    'Mean': mean,
    'Median': median,
    'Standard Deviation': std,
    'Interquartile Range': iqr
})

print(detailed_summary)
```

Positive Correlation: If the coefficient is close to +1, it means that as one variable increases, the other also increases.
Negative Correlation: If the coefficient is close to -1, it means that as one variable increases, the other decreases.
No Correlation: If the coefficient is around 0, it means there is no linear relationship between the variables.

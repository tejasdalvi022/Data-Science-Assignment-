```python
# Outliers are present in the following columns
outliers_columns = ['episodes','members']

# outlier_detection
def outlier_detection(data, col_name):
    Q1 = data[col_name].quantile(0.25)
    Q3 = data[col_name].quantile(0.75)
    IQR = Q3-Q1
    LE = Q1 - 1.5*IQR
    UE = Q3 + 1.5*IQR
    #return data.index[(data[col_name] < LE) | (data[col_name] > UE)].tolist()
    return LE,UE

# Dictionary for lower and upper extreme
LU_Extremes = {}
for col in outliers_columns:
    LU_Extremes[col] = outlier_detection(df, col)
LU_Extremes

# Outliers Capping
for col in LU_Extremes:
    df.loc[df[col] < LU_Extremes[col][0] , col]=LU_Extremes[col][0]
    df.loc[df[col] > LU_Extremes[col][1] , col]=LU_Extremes[col][1]
```

### Interview Questions:
#### Can you explain the difference between user-based and item-based collaborative filtering?
User-Based Collaborative Filtering Concept: Recommends items by identifying users with similar tastes.

Method: Looks for users who have rated items in a similar way and uses their preferences to suggest items.

Example: If User A and User B have similar ratings for several movies, and User B likes a new movie, that movie might be recommended to User A.

Pros: Good for small datasets, often more intuitive.

Cons: Struggles with scalability and sparsity when there are many users.

Item-Based Collaborative Filtering Concept: Recommends items by finding similarities between items.

Method: Looks for items that have been rated similarly and suggests them based on the user's past ratings.

Example: If you liked Movie X, and other users who liked Movie X also liked Movie Y, then Movie Y might be recommended to you.

Pros: Scales better for large datasets, handles sparsity better.

Cons: Requires a significant number of items to function well.

#### What is collaborative filtering, and how does it work?
Collaborative filtering is a popular technique used in recommendation systems. It’s based on the idea that people often get the best recommendations from someone with similar tastes and preferences.

How It Works User-Based Collaborative Filtering: Finds users who are similar to you based on their ratings of items. Recommends items that those similar users have liked but you haven’t tried yet.

Item-Based Collaborative Filtering: Looks at items that are similar to items you have liked. Recommends items that are similar to your previous likes based on the preferences of all users.

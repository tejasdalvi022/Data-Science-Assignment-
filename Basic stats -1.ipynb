# Standardization of Numerical Variables

###### Standardization, also known as z-score normalization, is a technique used to transform data so that it has a mean of 0 and a standard deviation of 1. This process adjusts the values of the data based on how far they deviate from the mean, measured in units of standard deviation. The formula for standardization is:(x – μ) / σ where:( x ) is the original value, ( \mu ) is the mean of the data, ( \sigma ) is the standard deviation of the data, ( z ) is the standardized value

# standarizing numerical columns

```python
# select numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
```

```python
# mean and standard deviation
mean = df[numerical_cols].mean()
std = df[numerical_cols].std()
# z-score normalisation 
df_standardized = (df[numerical_cols] - mean) / std
print("\nStandardized Data:")
print(df_standardized)
```

```python
# comaprison Before an after standardiztion
import matplotlib.pyplot as plt

# Plotting the original data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Data Distribution')
plt.hist(df['Volume'], bins=5, alpha=0.7, label='Volume')
plt.hist(df['Avg Price'], bins=5, alpha=0.7, label='Avg Price')
plt.hist(df['Discount Rate (%)'], bins=5, alpha=0.7, label='Discount Rate (%)')
plt.legend()

# Plotting the standardized data
plt.subplot(1, 2, 2)
plt.title('Standardized Data Distribution')
plt.hist(df_standardized['Volume'], bins=5, alpha=0.7, label='Volume')
plt.hist(df_standardized['Avg Price'], bins=5, alpha=0.7, label='Avg Price')
plt.hist(df_standardized['Discount Rate (%)'], bins=5, alpha=0.7, label='Discount Rate (%)')
plt.legend()

plt.show()
```

#### By standardizing the data, we ensure that each feature contributes equally to the analysis, improving the performance and accuracy of machine learning models.

# Conversion of Categorical Data into Dummy Variables

##### Why Convert Categorical Data into Dummy Variables?
Machine Learning Algorithms Require Numerical Input:
Most machine learning algorithms, especially those based on mathematical computations, require numerical input. Categorical data, which consists of labels or categories, needs to be converted into a numerical format to be processed by these algorithms.
Avoiding Ordinal Relationships:
Simply assigning numerical values to categories (e.g., 1 for “Red”, 2 for “Blue”, 3 for “Green”) can imply an ordinal relationship where none exists. One-hot encoding avoids this by creating separate binary columns for each category.
Improving Model Performance:
One-hot encoding can improve the performance of models by providing a clear and unambiguous representation of categorical variables. This helps the model to learn patterns more effectively.

```python
df_encoded = pd.get_dummies(df,categorical_columns)
print("\nOne-Hot Encoded Data:")
print(df_encoded)
```
